```python
import numpy as np
import os, glob
import matplotlib.pyplot as plt
import mplhep as hep
plt.style.use([hep.style.ROOT, hep.style.firamath])
import pickle
from matplotlib.colors import LinearSegmentedColormap
channel_list = ["Tracks_pt", "Tracks_dZSig", "Tracks_d0Sig", "ECAL_energy",
"HBHE_energy", "Pix_1", "Pix_2", "Pix_3", "Pix_4", "Tib_1", "Tib_2",
"Tib_3", "Tib_4", "Tob_1", "Tob_2", "Tob_3", "Tob_4", "Tob_5",
"Tob_6", "Tid_1", "Tec_1", "Tec_2", "Tec_3"]


# Define the CMS color scheme
cms_colors = [
    (0.00, '#FFFFFF'),  # White
    (0.33, '#005EB8'),  # Blue
    (0.66, '#FFDD00'),  # Yellow
    (1.00, '#FF0000')   # red
]

# Create the CMS colormap
cms_cmap = LinearSegmentedColormap.from_list('CMS', cms_colors)
```

```python
# U --> Tau Tau
```

```python
data_file = "../conversion_plot_scripts/plots_from_pq_unbiased_data_pkl/IMG_Upsilon_ToTauTau_Hadronic_tauDR0p4_m3p6To14_dataset_validation/data_for_validation_upsilonToTauTau_hadronic_dR0p4_a_jet_mass_pt.pkl"
infile = open(f"{data_file}", "rb")
data = pickle.load(infile)
print(f"loaded data file ------>  {data_file}")
infile.close()
jet_mass=data["jet_mass"]
a_mass=data["a_mass"]
a_pt=data["a_pt"]
jet_pt=data["jet_pt"]
mass_bins =np.arange(1.2, 14.4, .4)
pt_bins = np.arange(30,200,5)
fig, ax = plt.subplots(figsize=(20,15))
# norm = mcolors.TwoSlopeNorm(vmin=5000, vmax = 7000, vcenter=5500)
plt.hist2d(np.squeeze(a_mass), np.squeeze(a_pt), bins=[mass_bins, pt_bins],cmap=cms_cmap)
plt.xticks(np.arange(1.2, 14.4, .8))
plt.yticks(np.arange(10,200,10))
plt.xlabel(r'$\mathrm{mass}$ [GeV]')
plt.ylabel(r'$\mathrm{pT}$ [GeV]')
plt.colorbar().set_label(label='Events/ (0.4,5) GeV')
plt.grid(color='r', linestyle='--', linewidth=.2)
hep.cms.label(llabel="Simulation Preliminary", rlabel="13 TeV", loc=0, ax=ax)
```

```python
plt.hist(np.squeeze(a_mass), bins=mass_bins)
plt.hist(np.squeeze(jet_mass), bins=mass_bins,histtpe='step')
```
